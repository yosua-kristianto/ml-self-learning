# Nomor 2

Modul python ini berisi prosedur klasifikasi untuk data *mobile price* pada soal nomor 2.
https://www.kaggle.com/datasets/iabhishekofficial/mobile-price-classification

Algoritma Machine Learning yang akan digunakan pada klasifikasi ini akan dilakukan 2 kali dengan algoritma sebagai berikut:

1. Decision Tree
2. Random Forest

Data yang akan digunakan untuk melakukan machine learning ini adalah `test.csv` dan `train.csv`.

```python
# Importing required libraries

# Pandas is a tool to help dataset manipulation
import pandas

# Train test split
from sklearn.model_selection import train_test_split

# Decision Tree
from sklearn.tree import DecisionTreeClassifier

# Random Forest
from sklearn.ensemble import RandomForestClassifier

# MAE
from sklearn.metrics import mean_absolute_error
```

```python
# load dataset

dataset = pandas.read_csv('train.csv')

# Describe the dataset
dataset.head()
```

```python
# Defining feature & target
features = [
    'battery_power', 'blue', 'clock_speed',
    'dual_sim', 'fc', 'four_g', 'int_memory',
    'm_dep', 'mobile_wt', 'n_cores', 'pc', 'px_height',
    'px_width', 'ram', 'sc_h', 'sc_w', 'talk_time', 'three_g',
    'touch_screen', 'wifi'
]

predicition_target = dataset.price_range
```

## B-1. Decision Tree Section
This part, we using the training model that uses Decision Tree

```python
# Split data into training and validation data
train_X, val_X, train_y, val_y = train_test_split(dataset[features], predicition_target, random_state = 1)

# Defining Model
model_dt = DecisionTreeClassifier(random_state=1)
model_dt.fit(train_X, train_y)
```

```python
# Run test for predictions

print("Making prediction from train.csv")

# i only use 500 rows because the data since
test_dataset = pandas.read_csv('test.csv', nrows = 500)

test_dataset.head()

# Remove ID column
test_dataset = test_dataset.drop(columns=['id'])

# Describe the dataset
predicted_dt = model_dt.predict(test_dataset)
test_dataset['price_range'] = predicted_dt
pandas.set_option('display.max_rows', None)
test_dataset.head(None)
```

```python
# Performing MAE for Decision Tree

print(mean_absolute_error(val_y, predicted_dt))
```

## B-2. Random Forest Section
This part, we using the training model that uses Random Forest

```python
train_X, val_X, train_y, val_y = train_test_split(dataset[features], predicition_target, random_state = 1)

# Defining Model
model_rf = RandomForestClassifier(random_state=1)
model_rf.fit(train_X, train_y)

dataset[features].head()
```

```python
# Run test for predictions

print("Making prediction from train.csv")
test_dataset = pandas.read_csv('test.csv', nrows = 500)

test_dataset.head()

# Remove ID column
test_dataset = test_dataset.drop(columns=['id'])

# Describe the dataset
predicted_value_rf = model_rf.predict(test_dataset)
test_dataset['price_range'] = predicted_value_rf
pandas.set_option('display.max_rows', None)
test_dataset.head(None)
```

```python
# Performing MAE for Random Forest

print(mean_absolute_error(val_y, predicted_value_rf))
```
